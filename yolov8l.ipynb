```python
from ultralytics import YOLO


model = YOLO("yolov8l.pt") 

results = model.train(
    data="/kaggle/input/underwater-plastic-pollution-detection/underwater_plastics/data.yaml",  
    epochs=50,                  
    batch=16,                 
    imgsz=640,                  
    # device='cpu',                  
    name="custom_yolov8l_model", 
    workers=4                  
)
```
