```python
from ultralytics import YOLO

# Load a YOLOv8 model (e.g., YOLOv8n for Nano, YOLOv8s for Small)
model = YOLO("yolov8l.pt")  # Replace with yolov8s.pt, yolov8m.pt, etc., for larger models

# Train the model
results = model.train(
    data="/kaggle/input/underwater-plastic-pollution-detection/underwater_plastics/data.yaml",  # Path to your dataset config file
    epochs=100
    ,                  # Number of training epochs
    batch=16,                   # Batch size
    imgsz=640,                  # Image size for training
    # device='cpu',                   # GPU (use '0' for single GPU, or 'cpu' for CPU training)
    name="custom_yolov8l_model", # Name of the experiment
    workers=4                   # Number of dataloader workers
)
```
